```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import rc
from scipy import optimize, stats
rc('pgf', rcfonts = False, preamble = "\\usepackage{siunitx}\n\\usepackage[utf8]{inputenc}")
```

# Llei de Malus

Llei teòrica $I(\theta)=\frac{1}{2}\text{cos}^2(\theta)$

Intensitat en funció de l'angle de desviació.

* `I` intensitat mesurada per cada angle
* `Theta[i]` angle de gir del polaritzador

Angles: º
Intensitat: mW
Pecisió angles: 2º
Precisió intensitat: 0.01 mW

```python
Int=[0,0.06,0.15,0.30,0.46,0.65,0.79,0.91,0.99,0.95,0.93,0.97,0.98,1.00,1.00,1.01,1.01,1.00,1.00,0.99,0.98,0.97,0.96,0.94,0.92,0.90,0.88,0.86,0.72,0.56,0.38,0.21,0.09,0.02,0]
Theta=[80,90,100,110,120,130,140,150,160,154,152,156,158,162,164,166,168,170,172,174,176,178,180,182,184,186,188,190,200,210,220,230,240,250,260]
```

```python
def cosd(x):
    return np.cos(x * np.pi/180)

t = np.arange(Theta[0], Theta[-1], 0.1)
int_teo = 0.5* cosd(t)**2

theta_corr = [q - 167 for q in Theta]
```

```python
plt.figure(figsize = (5,3.5))
plt.grid(True)
plt.plot(theta_corr, Int, marker = '+', ms = 7, mfc = 'lightblue', ls = '')
plt.ylabel(r"\( I \, (\si{\mW}) \)")
plt.xlabel(r"\( \theta \, (\si{\degree}) \)")

plt.savefig('malus-1.pgf')
```

`Parells` és la llista dels punts a representar.

```python
cos = [cosd(q)**2 for q in theta_corr]
def model(x, A, B):
    return A + B*x
def model1(x,A,B):
    return A + B*cosd(x)**2
B, A, r, p, e = stats.linregress(cos, Int)
r**2, A, B
```

```python
params, params_err = optimize.curve_fit(model, cos, Int, p0 = [0,1])
np.sqrt(params_err[0][0]), np.sqrt(params_err[1][1])
```

```python
q = np.arange(0,1, 0.01)
plt.figure(figsize = (5,3.5))
plt.grid(True)
plt.plot(cos, Int, marker = '+', ms = 7, mfc = 'lightblue', ls = '')
plt.plot(q, model(q, A, B), ls = '--', color = 'orchid', lw = 1)
plt.ylabel(r"\( I \, (\si{\mW}) \)")
plt.xlabel(r"\( (\cos{\theta})^2 \)")

plt.savefig('regr.pgf')
```

```python
plt.figure(figsize = (5,3.5))
plt.grid(True)
plt.plot(theta_corr, Int, marker = '+', ms = 7, mfc = 'lightblue', ls = '')
q1 = np.arange(theta_corr[0], theta_corr[-1], 0.1)
plt.plot(q1, model1(q1, A, B), ls = '--', color = 'orchid', lw = 1)
plt.ylabel(r"\( I \, (\si{\mW}) \)")
plt.xlabel(r"\( \theta \, (\si{\degree}) \)")

plt.savefig('malus-2.pgf')
```
