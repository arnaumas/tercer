```python
import numpy as np
from matplotlib import rc
import matplotlib.pyplot as plt
```

```python
x,y = np.linspace(0,1.5,100),np.linspace(0,1.5,100)
X,Y = np.meshgrid(x,y)
a = 1
b = 1
U = X*(1 - X - a*Y)
V = Y*(1 - Y - b*X)
V1 = 0
```

```python
fig,ax = plt.subplots()

#('font', **{'family': 'sans-serif', 'serif': ['Computer Modern'], 'sans-serif': ['Computer Modern Sans serif']})
rc('text', usetex = True)
plt.rcParams['figure.figsize'] = [6,6]


ax.set_aspect('equal')
ax.set_ylim(0,1.5)
ax.set_xlim(0,1.5)
plt.tick_params(axis = 'both', which = 'both', bottom = False, top = False,\
                left = False, right = False, labelleft = False, labelbottom = False)
plt.plot([0,1], [1,0], color = 'green', linewidth = 5)

plt.plot(0,0, color = 'green', marker = 'o', markersize = 5, markeredgecolor = 'green', clip_on = False, zorder = 100)
plt.plot(1,0, color = 'green', marker = 'o', markersize = 5, markeredgecolor = 'green', clip_on = False, zorder = 100)
plt.plot(0,1, color = 'green', marker = 'o', markersize = 5, markeredgecolor = 'green', clip_on = False, zorder = 100)
plt.arrow(0,0.5, 0, 0, color = 'black', head_width = 0.05, head_length = 0.03, antialiased = True, linewidth = 0.5, width = 0.003, clip_on = False)
plt.arrow(0.5,0, 0.01, 0, color = 'black', head_width = 0.05,head_length = 0.03, antialiased = True,linewidth = 0.5, width = 0.003, head_starts_at_zero = True, clip_on = False)
plt.arrow(1.3,0, -0.01, 0, color = 'black', head_width = 0.05, head_length = 0.03,antialiased = True,linewidth = 0.5, width = 0.003, head_starts_at_zero = True, clip_on = False)
plt.arrow(0,1.3, 0, -0.01, color = 'black', head_width = 0.05, head_length = 0.03,antialiased = True,linewidth = 0.5, width = 0.003, head_starts_at_zero = True, clip_on = False)




plt.streamplot(x, y, U, V, color="orchid", linewidth = 0.8)
plt.savefig('retrat-6b.pdf', format = 'pdf', bbox_inches='tight')
plt.show()
```
